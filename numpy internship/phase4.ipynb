```python


from PIL import Image
from numpy import asarray
import numpy as np


# load the image and convert into 
# numpy array
img = Image.open('Sample.jpeg')
numpydata =asarray(img)

# data
print(numpydata)
np.save("img.npy",numpydata)
```

```python
# Display the image and its "dark" version
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(img)
plt.title("numpy img")
plt.grid(False)

# Convert PIL Image to numpy array for arithmetic
img_array = np.asarray(img)
filter_img = 130 - img_array  # invert colors for uint8 image

plt.subplot(122)
plt.imshow(filter_img)
plt.title("filter img")
```

```python
from PIL import Image #using pillow

# Open the image
img = Image.open("sample.jpeg")

# Convert to grayscale
bw_img = img.convert("RGB")

# Show the black and white image
bw_img.show()

# Save the black and white image
bw_img.save("Sample_bw.png")
```
