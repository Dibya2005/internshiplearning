datasets.data is your feature and dataset.target is your class

```python
df=pd.DataFrame(datasets.data,columns=datasets.feature_names)
df
```

```python
df.describe()
```

```python
x=df
y=datasets.target
```

```python
# iwould like to scale this feature before building my ml model
from sklearn.preprocessing import StandardScaler
sacler=StandardScaler()
X_scaled = sacler.fit_transform(x)
X_scaled # you get value -1 to 1
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

Use PCA to reduce dimensions

```python
x
```


Use components such that 95% of variance is retained

```python
from sklearn.decomposition import PCA

pca = PCA(0.95) # it will tell pca is that retain 95% of useful features 
X_pca = pca.fit_transform(x) # you supply x datafrasme got a new dataframe  with new feature new principle componants and when you do 
X_pca.shape # now you got only 29 columns it also calculate new column for me
```

```python
pca.explained_variance_ratio_
```

my 1 st column capturing 14% of variation or useful info from my dataset 2nd companant capturing 13% of info and so on

```python
pca.n_components_ # it will tell you how many componant you got
```

```python
# now good news is you can use this new dataframe to train your model
```

```python
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```


Let's now select only two components

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x)
X_pca.shape
```

```python
X_pca
```

```python

pca.explained_variance_ratio_
```


You can see that both combined retains 0.14+0.13=0.27 or 27% of important feature information

```python

X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test) # very low score but  benefit is that you havbe only two feature so your computation time is less 
# you can de3cide what accuracy do you want and you can reduce your dimmension accordingly
```


We get less accuancy (~60%) as using only 2 components did not retain much of the feature information. However in real life you will find many cases where using 2 or few PCA components can still give you a pretty good accuracy
