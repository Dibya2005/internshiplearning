## train the model and make prediction

```python
x=df.drop('target',axis=1)
x
```

```python
y=df.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
# model=RandomForestClassifier() by default n estimator is 10
#n estimator means no of tree in myb forest
model=RandomForestClassifier(n_estimators=50) # alg alg value dalke trai error krke usko dekh sakte ho
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

confusion matrics

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred=model.predict(X_test))
cm
```
