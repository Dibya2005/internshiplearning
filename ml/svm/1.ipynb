```python
import pandas as pd    
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names) # data is in iris.data and column is the feature name
df.head()
```

```python
df['taget']=iris.target # target column have the taget
df
```

```python
#posible value of target 0 , 1 ,2
iris.target_names
# 'setosa', 'versicolor', 'virginica' as 0,1,2
```

```python
df[df.taget==1].head()
```

```python
df[df.taget==2].head()
```

```python
df['flower_name'] = df.taget.apply(lambda x: iris.target_names[x]) #
#This applies a function to each value in the column 'taget' .apply aplly the following lembda function on each value of target 
df
```
