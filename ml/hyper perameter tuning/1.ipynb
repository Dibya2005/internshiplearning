```python
from sklearn import svm, datasets
from sklearn.model_selection import cross_val_score
import numpy as np

# Load data
iris = datasets.load_iris()

# Parameters
kernels = ['rbf', 'linear']
c_values = [1, 10, 20]
avg_score = {}

# Loop through combinations
#Combines the kernel and C into a key string like 'rbf_10'.

# Calculates the average score and stores it in the avg_score dictionary.
for kval in kernels:
    for cval in c_values:
        cv_score = cross_val_score(svm.SVC(kernel=kval, C=cval, gamma='auto'), iris.data, iris.target, cv=5)
        avg_score[kval + '_' + str(cval)] = np.average(cv_score)

# Print scores
print(avg_score)
```

```python
# loops are also not convinient
# beacuse the we have to do too many iteration
# sklearn provide an api called grid search cv
# which will do the exact same thing 
```

to tackle this kind of priobklem  sklearn introduce a concept called rendomised cv


```python
# it will not try every permutation and combinationa of perameter it will try rendom combination of this perameter
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(svm.SVC(gamma='auto'), {
    'C': [1, 10, 20],
    'kernel': ['rbf', 'linear']
},
cv=5, return_train_score=False,
n_iter=2 # we can chose how many perameter you want to do
)
rs.fit(iris.data, iris.target)
pd.DataFrame(rs.cv_results_)[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
from sklearn import svm 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
modelperms = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params': {
            'C': [1, 10, 20],
            'kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'n_estimators': [1, 5, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'C': [1, 5, 10]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV

score = []
for model_name, mp in modelperms.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    score.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_peram': clf.best_params_
    })
```

```python
df = pd.DataFrame(score, columns=['model', 'best_score', 'best_peram'])
df
```
